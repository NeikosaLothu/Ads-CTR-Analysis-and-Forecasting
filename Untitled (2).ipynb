```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from datetime import timedelta, datetime
```

```python

np.random.seed(42)


start_date = datetime.strptime("2024-01-01", "%Y-%m-%d")
days = 60
dates = [start_date + timedelta(days=i) for i in range(days)]


ad_ids = ["AD001", "AD002", "AD003"]
platforms = ["Facebook", "Instagram", "Google"]
devices = ["Mobile", "Desktop", "Tablet"]

data = []

for date in dates:
    for ad_id in ad_ids:
        platform = random.choice(platforms)
        device = random.choice(devices)
        impressions = np.random.randint(800, 2000)
        clicks = np.random.binomial(impressions, p=0.03 + 0.01 * random.random())
        data.append([date.strftime('%Y-%m-%d'), ad_id, platform, device, impressions, clicks])

df_ads = pd.DataFrame(data, columns=["date", "ad_id", "platform", "device_type", "impressions", "clicks"])


df_ads.to_csv("ads_data.csv", index=False)
print("Dataset generated and saved as 'ads_data.csv'")
```

```python

df = pd.read_csv("ads_data.csv")


df['date'] = pd.to_datetime(df['date'])


df.dropna(inplace=True)


df = df[df['impressions'] > 0]


df.reset_index(drop=True, inplace=True)


print("Cleaned DataFrame info:")
print(df.info())
df.head()
```

```python

df_grouped = df.groupby(['date', 'ad_id']).agg({
    'impressions': 'sum',
    'clicks': 'sum'
}).reset_index()


df_grouped['ctr'] = df_grouped['clicks'] / df_grouped['impressions']


df_grouped.sort_values(by='date', inplace=True)

df_grouped.head()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))


for ad in df_grouped['ad_id'].unique():
    ad_data = df_grouped[df_grouped['ad_id'] == ad]
    plt.plot(ad_data['date'], ad_data['ctr'], marker='o', label=ad)


plt.title("CTR Over Time by Ad ID", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("CTR (Click-Through Rate)", fontsize=12)
plt.legend(title="Ad ID")
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)

plt.show()
```

```python

platform_ctr = df.groupby('platform').agg({
    'impressions': 'sum',
    'clicks': 'sum'
}).reset_index()

platform_ctr['ctr'] = platform_ctr['clicks'] / platform_ctr['impressions']


plt.figure(figsize=(8, 5))
plt.bar(platform_ctr['platform'], platform_ctr['ctr'], color=['#1f77b4', '#ff7f0e', '#2ca02c'])

plt.title('Average CTR by Platform', fontsize=14)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Average CTR', fontsize=12)
plt.ylim(0, platform_ctr['ctr'].max() * 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(platform_ctr['ctr']):
    plt.text(i, v + 0.001, f"{v:.3f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
